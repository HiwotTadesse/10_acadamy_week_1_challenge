```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/raw_analyst_ratings.csv')
```

```python
df['converted_date'] = df["date"].str.split(" ")
```

```python
df['converted_date'] = pd.to_datetime(df['converted_date'].str[0], errors='coerce', format='%Y-%m-%d %H:%M:%S')
```

```python
publication_counts = df.groupby(df['converted_date'].dt.date).size()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(publication_counts.index, publication_counts.values)
plt.xlabel('Date')
plt.ylabel('Publication Frequency')
plt.title('Publication Frequency Over Time')
plt.xticks(rotation=45)
plt.show()
```
